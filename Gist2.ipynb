```python
import pandas as pd

data = {'Alpha':  [1,5,9,6,5,8,4,7,5,1],
        'Beta':  [1,5,5,5,2,1,2,2,1,1],
        'Gamma':  [665,9985,2,46,6,6,77,8,53,54],
        'Delta':  [9,9,5,2,5,6,5,5,7,4],
        'Echo' : [223,4,343,9,566,545,7,5,44,1]
        }

df = pd.DataFrame (data, columns = ['Alpha','Beta','Gamma','Delta','Echo'])

print (df)
```

```python
df.groupby('Alpha').mean()
```

```python
df.groupby(['Alpha','Beta']).mean()
```

```python
df.groupby(['Alpha','Beta'], as_index=False).mean()
```

```python
df.groupby(['Alpha','Beta'], as_index=False)['Delta'].mean()
```
